```
import urllib
import pandas
import matplotlib.pyplot as p

def main():
	url = "https://raw.githubusercontent.com/datasets/gdp/0be54c18d900edc37123f25b4eff014731c9e459/data/gdp.csv"
	file = urllib.request.urlopen(url)
	data = pandas.read_csv(file)
	data['Value'] = data['Value'].divide(1000000000)

	US = data[data['Country Name'].str.match("United States")]
	EU = data[data['Country Name'].str.match("European Union")]
	China = data[data['Country Name'].str.match("China")]

	p.plot(US['Year'], US['Value'], '--', label='United States', linewidth=2)
	p.plot(EU['Year'], EU['Value'], ':', label='European Union', linewidth=2)
	p.plot(China['Year'], China['Value'], '-.g', label='China', linewidth=2)
	p.title("Conor Logsdon's GDP Plot")
	p.xlabel('Year')
	p.ylabel('GDP (Billions)')
	p.legend()
	p.show()
	p.close()


main()

```
